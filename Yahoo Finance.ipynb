```python
out = widgets.Output(layout={'border': '0.1px solid black', 'width': '1200px'})
symbol = pn.widgets.Select(name='Choose your symbol to get tranding news', options=['APD', 'APH', 'RACE', 'TEL', 'IFF', 'GLW'])

#@out.capture()
@pn.depends(symbol)
def get_news(symbol):
    get_new = yf_get_news(yf_get_ticker(symbol))
    out.clear_output(wait=False)
    with out:
        for i in range(1, len(get_new)):
            display(get_new[i]['title'])
            display(get_new[i]['link'] )
    return out
                                                                                    
def yf_get_ticker(symbol):
    yf_ticker = yf.Ticker(symbol)
    return yf_ticker

def yf_get_news(symbol):
    new = symbol.news
    return new


pn.Row(    
      pn.Column('#### Select a symbol from the list', 
                symbol, 
                get_news)
)



```
