```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import ipywidgets as widgets
import numpy as np
import panel as pn                                      #Librariy for panels
pn.extension()                                          #Load panel extension
```

```python
def crt(out1, out2):
    tab = widgets.Tab(children = [out1, out2])
    tab.set_title(0, 'First')
    tab.set_title(1, 'Second')
    #display(tab)
    return tab

def plot1():
    out1 = widgets.Output()
    data1 = pd.DataFrame(np.random.normal(size = 50))
    
    with out1:
        fig1, axes1 = plt.subplots()
        data1.hist(ax = axes1)
        plt.show(fig1)
    return out1
        
def plot2():
    out2 = widgets.Output()
    data2 = pd.DataFrame(np.random.normal(size = 100))
    
    with out2:
        fig2, axes2 = plt.subplots()
        data2.hist(ax = axes2)
        plt.show(fig2)
        
    return out2
```

```python
crt(plot1(), plot2())
```

```python
layout = pn.Column(crt(plot1(), plot2()), height=300, width=600)
```

```python
title = pn.pane.Markdown(""" 
        ### Responsible Investor """,  
        width=900)


tabs = pn.Tabs(
    ("Plots", 
       pn.Column(layout , height=400, width=200)),
    
    ("Plots2", 
       pn.Column(layout , height=400, width=200))
)

tabs
```
